# ANN- Regression analysis on **Combined Cycle Power Plant Data Set**

## Importing Libraries

```
import numpy as np
import pandas as pd
import tensorflow as tf
```

## Data preprocessing

### Importing data set

```
df=pd.read_excel('Folds5x2_pp.xlsx')
```

```
df.head()
```

```
df.isnull().sum()
```

```
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

### Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

## Building ANN

### Initializing ANN

```
model=tf.keras.Sequential()
```

### Adding the input layer and the first hidden layer

```
model.add(tf.keras.layers.Dense(6,activation='relu'))
```

### Adding the second hidden layer

```
model.add(tf.keras.layers.Dense(6,activation='relu'))
```

### Adding the output layer

```
model.add(tf.keras.layers.Dense(1))
```

Note: we did not add any activation function to the output layer because we need continuous value after the weighted sum

## Part 3 - Training the ANN

### Compiling the ANN

```
model.compile(optimizer="adam",loss='mean_squared_error')
```

### Training the ANN model on the Training set

```
model.fit(x_train,y_train,epochs=150,batch_size=30)
```

## Predicting the results of the Test set

```
y_pred = model.predict(x_test)
np.set_printoptions(precision=2)
y_pred
```

```
y_test=np.array(y_test)
```

```
y_test=y_test.reshape(len(y_test),1)
```

```
result=np.concatenate((y_pred,y_test), axis=1)
```

```
result
```

```
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
r2
```

r2 score is almost close to 1. we can say our model is a perfect model
